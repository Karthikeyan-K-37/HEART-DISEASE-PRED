```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv('data.csv')
heart_data
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
heart_data.isnull().sum()
```

```python
heart_data['Heart Disease'].value_counts()
```

```python
x=heart_data.drop(columns='Heart Disease',axis=1)
y=heart_data['Heart Disease']
x
```

```python
y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.5,stratify=y,random_state=2)
```

```python
X_train
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
accure=accuracy_score(X_train_prediction,Y_train)
```

```python
accure
```

```python
X_test_prediction=model.predict(X_test)
accure1=accuracy_score(X_test_prediction,Y_test)
```

```python
accure1
```
